Abaixo é realizado o import de "matplotlib" que serve para construção dos graficos a partir dos dados tratados, "pandas" para a analise e manipulação dos dados e o "numpy" para trabalhar com arrays e matrizes

```python
import matplotlib as mp
import pandas as pd
import numpy as py
```

Aqui é feita a leitura dos dados do csv "tv_shows"


```python
tvShows = pd.read_csv("raw/tv_shows.csv")
tvShows
```

Abaixo ocorre uma limpeza dos dados de "tvShows" atraves do ".dropna" que retira quaisquer linhas e colunas que tenham valores ausentes (NaN/None)

```python
tvShowsClean = tvShows.dropna()
```

Aqui está sendo feita a filtragem dos dados das notas dos shows de tv no IMDb e no Rotten Tomatoes e uma alteração nos dados para que as notas apareçam de 0 a 100 em ambas as colunas.

```python
tvShowsClean['imdb_100'] = pd.to_numeric(
    tvShowsClean['IMDb'].astype(str).str.replace('/10', ''),
    errors='coerce'
) * 10
tvShowsClean['rottenTomatoes_100'] = pd.to_numeric(
    tvShowsClean['Rotten Tomatoes'].astype(str).str.replace('/100', ''),
    errors='coerce'
)
```

Aqui padronizamos os valores das notas do IMDb e do Rotten Tomatoes para float.

```python
tvShowsClean['imdb_100'] = tvShowsClean['imdb_100'].astype(float)
tvShowsClean['rottenTomatoes_100'] = tvShowsClean['rottenTomatoes_100'].astype(float)
```

Por termos acima as colunas de IMDb e Rotten Tomatoes já criadas com os valores filtrados e alterados de maneira que as notas fiquem padronizadas fazemos abaixo o drop (descarte) das colunas com os dados não filtrados para não ter duplicatas e com a "unnamed: 0" descartamos qualquer coluna sem nome no CSV

```python
df_tvShows = tvShowsClean.drop(columns=['IMDb', 'Rotten Tomatoes', 'Type', 'Unnamed: 0'])
```

Na linha abaixo fazemos uma filtragem onde para cada show de tv haverá uma indicação de sim e não caso aquele show esteja disponível ou não para um serviço de streaming especifico

```python
## df_tvShows.loc[:, colunas] = df_tvShows[colunas].replace({1: 'SIM', 0: 'NÃO'})

colunas = [c for c in df_tvShows.columns if c in ['Netflix', 'Hulu', 'Prime Video', 'Disney+']]
for c in colunas:
    df_tvShows[c] = df_tvShows[c].map({1: 'SIM', 0: 'NÃO'}).fillna(df_tvShows[c]).astype(str)
```

Aqui filtramos todos os shows que foram lançados depois dos anos 2000


```python
df_tvShows_2000 = df_tvShows[tvShowsClean['Year'] >= 2000]
```

Abaixo pegamos os shows: e filtramos aqueles que tem nota maior que 70 no rotten tomatoes

```python
df_tvShows_rotten_70_plus = df_tvShows_2000[df_tvShows_2000['rottenTomatoes_100'] >= 70]
```

Abaixo pegamos os shows: e filtramos aqueles que tem nota maior que 70 no Rotten tomatoes e no IMDb



```python
df_tvShows_rotten_imdb_70_plus = df_tvShows_rotten_70_plus[df_tvShows_rotten_70_plus['imdb_100'] >= 70]
```

Aqui pegamos todos os shows lançados depois dos anos 2000 e filtramos apenas aqueles que são recomendados para "all"(todos), 7+(anos) e 13+(anos) e colocamos na variável "publico_infantil"



---



```python
publico_infantil = df_tvShows_rotten_imdb_70_plus[tvShowsClean['Age'].isin(['all', '7+', '13+'])]
```

Aqui pegamos todos os shows lançados depois dos anos 2000 e filtramos apenas aqueles que são recomendados para 16+(anos) e 18+(anos) e colocamos na variável "publico_adulto"

```python
publico_adulto = df_tvShows_rotten_imdb_70_plus[tvShowsClean['Age'].isin(['16+', '18+'])]
```
